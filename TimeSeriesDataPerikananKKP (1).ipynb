```python
# split data
train_size = int(len(tangkaplautkakap) * 0.7) # Menentukan banyaknya data train yaitu sebesar 70% data
train = tangkaplautkakap[:train_size]
test = tangkaplautkakap[train_size:].reset_index(drop=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(train[['Volume Produksi']])

train['scaled'] = scaler.transform(train[['Volume Produksi']])
test['scaled'] = scaler.transform(test[['Volume Produksi']])
```

```python
train.head()
```

```python
def sliding_window(data, window_size):
    sub_seq, next_values = [], []
    for i in range(len(data)-window_size):
        sub_seq.append(data[i:i+window_size])
        next_values.append(data[i+window_size])
    X = np.stack(sub_seq)
    y = np.array(next_values)
    return X,y
```

```python
window_size = 1

X_train, y_train = sliding_window(train[['scaled']].values, window_size)
X_test, y_test = sliding_window(test[['scaled']].values, window_size)
```

```python
#input LSTM must 3 variable
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM #, RNN, GRU 
```

```python
def create_model(LSTM_unit=64, dropout=0.2): #jika ingin menggunakan RNN atau GRU ganti LSTM dengan GRU/RNN
    # create model
    model = Sequential()
    model.add(LSTM(units=LSTM_unit, input_shape=(window_size, 1)))
    model.add(Dense(32, activation='relu'))
    model.add(Dropout(dropout))
    model.add(Dense(1))
    # Compile model
    model.compile(loss='mse', optimizer='adam', metrics=['mae'])
    return model
```

```python
LSTM_unit = [16,32,64,128]
dropout = [0.1,0.2]
```

```python
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasRegressor
from keras.callbacks import EarlyStopping
# Early Stopping
es = EarlyStopping(monitor = 'val_loss', mode = "min", patience = 5, verbose = 1)

# create model
model = KerasRegressor(build_fn=create_model, epochs=50, validation_split=0.1, batch_size=1, callbacks=[es], verbose=1)

# define the grid search parameters
LSTM_unit = [16,32,64,128]
dropout=[0.1,0.2]
param_grid = dict(LSTM_unit=LSTM_unit, dropout=dropout)
```

```python
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=5)
```

```python
grid_result = grid.fit(X_train, y_train)
```
